```python
import pandas as pd
import numpy as np
```

```python
s1=pd.Series(np.arange(4),
             index=["a","c","d","e"])
s2=pd.Series(np.arange(5),
             index=["a","c","e","f","g"])
```

```python
print(s1)
print(s2)
```

```python
s1+s2
```

```python
df1=pd.DataFrame(
    np.arange(6).reshape(2,3),
    columns=list("ABC"),
    index=["Tim","Tom"])
df2=pd.DataFrame(
    np.arange(9).reshape(3,3),
    columns=list("ACD"),
    index=["Tim","Kate","Tom"])
```

```python
print(df1)
print(df2)
```

```python
df1+df2
```

```python
df2.add(df1,fill_value=0)
```

```python
1/df1
```

```python
df1*3
```

```python
df1.mul(3)
```

```python
df2
```

```python
s=df2.iloc[1]
s
```

```python
s2=df2["A"];
s2
```

## Applying a Function

```python
df=pd.DataFrame(
    np.random.randn(4,3),
    columns=list("ABC"),
    index=["Kim","Susan","Tim","Tom"])
df
```

```python
np.abs(df)
```

```python
df
```

```python
f=lambda x:x.max()-x.min()
```

```python
df.apply(f,axis=0)
```

```python
df.apply(f,axis=1)
```

```python
def f(x):
    return x**2
```

```python
df.iloc[0].apply(f)
```
