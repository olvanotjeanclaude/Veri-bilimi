```python
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame(
    [[2.4,np.nan],[6.3,-5.4],
     [np.nan,np.nan],[0.75,-1.3]],
    index=["a","b","c","d"],
    columns=["one","two"])
df
```

```python
df.sum()
```

```python
df.sum(axis=1)
```

```python
df.mean(axis=1)
```

```python
df.mean(axis=1,skipna=False)
```

```python
df.idxmax()
```

```python
df.idxmin()
```

```python
df.cumsum()
```

```python
df.describe()
```

To find the correlation coefficient, let's first import the famous iris data set. You can download iris data set from https://archive.ics.uci.edu/ml/datasets/iris.

```python
iris=pd.read_csv("iris.data",
                 sep=",",
                 header=None)
```

```python
iris
```

```python
iris.columns=['sepal_length','sepal_width',
             'petal_length','petal_width',
             'class']
```

```python
iris.head()
```

```python
iris["sepal_length"].corr(iris["sepal_width"])
```

```python
iris.corr()
```

```python
iris.cov()
```

```python
iris.corrwith(iris.petal_length)
```
